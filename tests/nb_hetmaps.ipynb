```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load numpy matrices
knn_multi = np.load("parallel/multi_knn_matrix.npy")
kde_multi = np.load("parallel/multi_kde_matrix.npy")

knn_multi = knn_multi + knn_multi.T + np.diag(np.ones(knn_multi.shape[0]))
kde_multi = kde_multi + kde_multi.T + np.diag(np.ones(kde_multi.shape[0]))

# Plots
fig, (ax1, ax2) = plt.subplots(figsize=(13, 5), ncols=2)

plt1 = ax1.imshow(knn_multi, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt1, ax=ax1)
ax1.set_title("Generalized Correlation Matrix via KNN")

plt2 = ax2.imshow(kde_multi, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt2, ax=ax2)
ax2.set_title("Generalized Correlation Matrix via KDE")

plt.suptitle("Generalized Correlation Matrix", fontsize=15)
plt.show()
```

```python
# Load numpy matrices
lmi = np.load("linear/linear_info_matrix_new.npy")
pea = np.load("linear/linear_info_matrix.npy")

lmi = lmi + lmi.T + np.diag(np.ones(lmi.shape[0]))
pea = pea + pea.T + np.diag(np.ones(pea.shape[0]))

# Plots
fig, (ax1, ax2) = plt.subplots(figsize=(13, 5), ncols=2)

plt1 = ax1.imshow(lmi, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt1, ax=ax1)
ax1.set_title("Via Linear Mutual Information")

plt2 = ax2.imshow(pea, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt2, ax=ax2)
ax2.set_title("Generalized Correlation Matrix via KDE")

plt.suptitle("Correlation Matrix", fontsize=15)
plt.show()
```
