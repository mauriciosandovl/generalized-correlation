```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load numpy matrices
kde_vect = np.load("vectorial/vect_kde_matrix.npy")
kde_multi = np.load("parallel/multi_kde_matrix.npy")

kde_vect = kde_vect + kde_vect.T + np.diag(np.ones(kde_vect.shape[0]))
kde_multi = kde_multi + kde_multi.T + np.diag(np.ones(kde_multi.shape[0]))

# Plots
fig, (ax1, ax2) = plt.subplots(figsize=(13, 5), ncols=2)

plt1 = ax1.imshow(kde_vect, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt1, ax=ax1)
ax1.set_title("Via KDE")

plt2 = ax2.imshow(kde_multi, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt2, ax=ax2)
ax2.set_title("Via parallel KDE")

plt.suptitle("Generalized Correlation Matrix", fontsize=15)
plt.show()
```

```python
# Load numpy matrices
knn_vect = np.load("vectorial/vect_knn_matrix.npy")
knn_multi = np.load("parallel/multi_knn_matrix.npy")

knn_vect = knn_vect + knn_vect.T + np.diag(np.ones(knn_vect.shape[0]))
knn_multi = knn_multi + knn_multi.T + np.diag(np.ones(knn_multi.shape[0]))

# Plots
fig, (ax1, ax2) = plt.subplots(figsize=(13, 5), ncols=2)

plt1 = ax1.imshow(knn_vect, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt1, ax=ax1)
ax1.set_title("Via KNN")

plt2 = ax2.imshow(knn_multi, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt2, ax=ax2)
ax2.set_title("Via parallel KNN")

plt.suptitle("Generalized Correlation Matrix", fontsize=15)
plt.show()
```

```python
# Load numpy matrices
lmi = np.load("linear/linear_info_matrix.npy")
pea = np.load("linear/pearson_matrix.npy")

lmi = lmi + lmi.T + np.diag(np.ones(lmi.shape[0]))
pea = pea + pea.T + np.diag(np.ones(pea.shape[0]))

# Plots
fig, (ax1, ax2) = plt.subplots(figsize=(13, 5), ncols=2)

plt1 = ax1.imshow(pea, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt1, ax=ax1)
ax1.set_title("Via Pearson")

plt2 = ax2.imshow(lmi, cmap="jet", interpolation="bilinear", origin="lower")
fig.colorbar(plt2, ax=ax2)
ax2.set_title("Via Linear Mutual Information")

plt.suptitle("Correlation Matrix", fontsize=15)
plt.show()
```
